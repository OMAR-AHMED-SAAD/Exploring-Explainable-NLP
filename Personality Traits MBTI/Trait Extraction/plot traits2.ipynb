```python
import matplotlib.pyplot as plt
import json
```

```python
# load traits percentages
with open('Bill Gates traits.json') as f:
    gates_traits = json.load(f)

with open('Elon Musk traits.json') as f:
    musk_traits = json.load(f)

with open('Jack Dorsey traits.json') as f:
    dorsey_traits = json.load(f)

with open('Jeff Bezos traits.json') as f:
    bezos_traits = json.load(f)

with open('Mark Zuckerberg traits.json') as f:
    zuckerberg_traits = json.load(f)

with open('Tim Cook traits.json') as f:
    cook_traits = json.load(f)  
```

```python

def plot(percentages, labels, names):
    fig, axs = plt.subplots(3, 2, figsize=(10,10))
    axs = axs.flatten()
    fig.suptitle('Personality Traits Distribution')
    
    for i, percentage in enumerate(percentages):
        # Round the percentages to two decimal places
        # Plotting
        zeros_percentages=[percentage["Introversion"],percentage["Intuition"],percentage["Thinking"],percentage["Judging"]]
        ones_percentages=[percentage["Extroversion"],percentage["Sensing"],percentage["Feeling"],percentage["Perceiving"]]
        ax = axs[i]
        ax.bar(labels, zeros_percentages, label='(0)')
        ax.bar(labels, ones_percentages, bottom=zeros_percentages, label='(1)')
        ax.set_ylim([0, 100])  # Set y-axis limit from 0 to 1
        ax.set_ylabel('Percentage')
        ax.set_title(names[i])
        ax.legend()
        # flip the x-axis labels
        
    
    plt.tight_layout()
    plt.show()

# Example usage
percentages = [musk_traits, gates_traits, dorsey_traits, bezos_traits, zuckerberg_traits, cook_traits]
labels = ['I(0) vs E(1)', 'N(0) vs S(1)', 'T(0) vs F(1)', 'J(0) vs P(1)']
plot(percentages, labels,['Elon Musk','Bill Gates','Jack Dorsey','Jeff Bezos','Mark Zuckerberg','Tim Cook'])
```
